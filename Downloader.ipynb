```python
import pandas as pd

# Load the CSV file into a DataFrame
dataset = pd.read_csv('Documentlist.csv',sep=';')

# Display the first few rows of the dataset
dataset.head()
```

```python
import shutil

for i in range(0, len(dataset)):
    if dataset['OE'][i] == 'BOA':
        if shutil.os.path.exists('Downloads/' + str(i)+ '.pdf') and not shutil.os.path.exists('Downloads_BOA/'+ dataset['Titel'][i] +'.pdf'):
            if dataset['Gültigkeit'][i].endswith('-'):
                shutil.copy('Downloads/' + str(i)+ '.pdf', 'Downloads_BOA/'+ dataset['Titel'][i] +'.pdf')
```

```python
import requests
import os

# Iterate through the dataset and download the files
for index, row in dataset.iterrows():
    url = row['URL']
    file_name = 'Downloads/' + str(index) + '.pdf'

    cookies = {'AL_SESS_IXOP-S': 'ATQC!CxIV89d4UXbbzGKXHiOzbBCkoKNvwWmCeQ1u!8wHT9jkgleeHjd0yvYhp_wWmaj'}
    
    

    if os.path.exists(file_name):
        print(f"{file_name} already exists. Skipping download.")
        continue
    
    if not url.startswith('http'):
        print(f"No URL given for {file_name}")
        continue
    
    response = requests.get(url, cookies=cookies)
    if not response.headers['Content-Disposition'].endswith('.pdf"'):
        print(response.headers['Content-Disposition'])
    if response.status_code == 200 and response.headers['Content-Type'] == 'application/octet-stream' and response.headers['Content-Disposition'].endswith('.pdf"'):
        with open(file_name, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {file_name}")
    else:
        print(f"Failed to download {file_name}")
```
